### <span style="color:purple"><div style="text-align: justify">We decided to study the impact of traumatic events on the number of quotes related to the metoo movement</div></span>

### <span style="color:red"><div style="text-align: justify">As traumatic events we are using the reporting of sexual harassment or assault-allegations against celebrities, politicians, CEOs, and others. To do so, we scrape the dates from the following url: "https://www.vox.com/a/sexual-harassment-assault-allegations-list?__c=1"</div></span>

```python
# To do so, we scrape the dates from the following url
URL = "https://www.vox.com/a/sexual-harassment-assault-allegations-list?__c=1"

# Scrape the text of this URL
html = urlopen(URL).read()
soup = BeautifulSoup(html, features="html.parser")
```

```python
#Kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

#Get text
text = soup.get_text()
```

```python
#Break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
#Break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
#Drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)
```

### <span style="color:purple"><div style="text-align: justify">Converting the soup of text into a readable dataframe</div></span>

```python
# function to parse the html
# htmlmarkers are used to separate to separate the soup of text into the different events: 
# the htmlmarker actually corresponds to the number of events per catergory: "Arts & Entertainment, Media, Business & Tech, Politics, Other" (see website)
htmlmarkers = ['/ 57', '/ 18', '/ 46','/ 40']

# initial = text
accused_raw = text.split('/ 101')
accused_all = []
i = 0

for marker in htmlmarkers:
  
  accused_clean = accused_raw[1:-1]
  accused_all.append(accused_clean)

  accused_new_raw = accused_raw[-1].split(marker) #splitting the rest of the page
  accused_raw = np.copy(accused_new_raw)

accused_clean = accused_raw[1:-1]
accused_all.append(accused_clean)
```

```python
# Now lets put all these events into a readable and usable dataframe
#  Generate dataframe of events, dates names and categories

cats = ['Arts & Entertainment','Media','Business & Tech','Politics', 'Other'] # These are the categories to which the accused people belong


events = pd.DataFrame()
for c,accused in enumerate(accused_all):
    
    names = [string.split('\n')[3] for string in accused] # get the names
    names[0] = accused[0].split('\n')[1] #the first one is formatted differently

    dates = [string.split('\n')[6] if string.split('\n')[6][-1].isdigit() else (string.split('\n')[6] +' '+ string.split('\n')[7]) for string in accused]

    dates[0] = accused[0].split('\n')[4] #the first one is formatted differently

    cat = cats[c] 
    cat = [cat] * len(names) 

    events_tmp = pd.DataFrame({'dates': dates,'names': names,'category': cat}) 

    events = events.append(events_tmp,ignore_index = True)

#dropping the event which did not have a correct date attached to it
events.drop(events[events.dates.str.isalpha()].index, inplace=True)

# What our dataframe with traumatic events looks like
events
```

```python
events['dates']=  pd.to_datetime(events['dates'], infer_datetime_format=True) # Again the dates into datetime_format
```

## <span style="color:Blue">4) Quantifying the impact of traumatic events</span>

### <span style="color:purple"><div style="text-align: justify">Now to get an idea of which events actually had an impact on the number of quotes mentionning metoo we are going to quantify it.</div></span>

```python
def impact(quotes, events, res = 'day', method = 'mean-mean', window_before = 7,window_after = 7):
  '''
  input 
  df:             quotes timeline
  events:         events (name, field, date)
  res:            resolution, interval of quotes taken together day, month
  method:         mean-mean, mean-max, diffpeak
  window_before:  points before for mean based methods
  window_after:   points after

  output
  impact_vect:    list of impacts for all events
  '''

  impact_vect = []
  window_before = pd.Timedelta(window_before, unit = res)
  window_after = pd.Timedelta(window_after, unit = res)
  
  for date in events['dates']:
    '''take the values before and after by slicing in the right dates'''

    if method == 'mean-mean':
      
      mean_before = quotes[(quotes[res] < date) & (quotes[res] > date - window_before)]['quotation'].mean()
      mean_after = quotes[(quotes[res] < date + window_before) & (quotes[res] > date) ]['quotation'].mean()
     
      impact = mean_after - mean_before

    if method == 'mean-max':
      mean_before = quotes[(quotes[res] < date) & (quotes[res] > date - window_before)]['quotation'].mean()
      max_after = quotes[(quotes[res] < date + window_before) & (quotes[res] > date)]['quotation'].max()
      impact = max_after - mean_before


    if method == 'diff':
      quotes['grad'] = quotes['quotation'].diff()/quotes[res].diff().dt.days()
      print(quotes['grad'])     
      impact = mean_after - mean_before
    
    
    
    impact_vect.append(impact)


  return impact_vect
```

```python
# Time window over which we measure the impact of our events
window_impact = 7
# Measuring the impact for all traumatic events (sexual assault and haracement accusation)
# Using the min-max method
impact_vect_max = impact(quotes_day, events, res = 'day', method = 'mean-max', window_before = window_impact,window_after = window_impact)
# Using the difference in means of quotes before and after to quantify the impact
window_impact = 14
impact_vect_mean = impact(quotes_day, events, res = 'day', method = 'mean-mean', window_before = window_impact,window_after = window_impact)
# Adding the impacts associated to the events to the dataframe
events['impact_max'] = impact_vect_max
events['impact_mean'] = impact_vect_mean
# Sorting the events based on impact
events.sort_values('impact_mean', inplace = True, ascending = False)
events.head(n = 15)
```
